```python
from sklearn.datasets import load_iris
data=load_iris()
```

```python
X=data['data']
y=data['target']
```

```python
import numpy as np
X_features=X[:,2:3]
y_features=(y==2).astype(np.uint8)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.svm import LinearSVC
```

```python
from sklearn.preprocessing import StandardScaler
model=Pipeline([
    ('StandScaler',StandardScaler()),
    ('SVM',LinearSVC(C=100,loss='hinge'))
])
```

```python
model.fit(X_features,y_features)
```

```python
from sklearn.metrics import accuracy_score
y_predict=model.predict(X_features)
print(accuracy_score(y_features,y_predict))
```

```python
from sklearn.preprocessing import PolynomialFeatures
model=Pipeline(
    [
        ('Poly',PolynomialFeatures(degree=3)),
        ('StandScaler',StandardScaler()),
        ('SVM',LinearSVC(C=100,loss='hinge'))
    ]
)
```

```python
model.fit(X,y)
```

```python
import matplotlib.pyplot as plt
plt.subplot(1, 2, 1)
plt.scatter(X[:,0],X[:,1],c=y,cmap='jet')
plt.subplot(1, 2, 2)
plt.scatter(X[:,0],X[:,1],c=model.predict(X),cmap='jet')
plt.show()
```

```python
X=np.random.uniform(-5, 5,(100,1))
noise=np.random.randint(1,10,(100,1))
y=5*X+3*X*X+np.random.rand(100,1)+noise
```

```python
plt.scatter(X,y)
```

```python
from sklearn.svm import SVR
model=SVR(kernel='poly',degree=2,C=100,epsilon=0.1)
model.fit(X,y)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.svm import SVR

# Assuming X and y are already defined (from your previous code)

model = SVR(kernel='poly', degree=2, C=100, epsilon=0.1)
model.fit(X, y)  # Reshape y to 1D if necessary: model.fit(X, y.ravel())

# Create a range of x values for plotting
x_range = np.arange(X.min(), X.max(), 0.1)[:, np.newaxis]

# Predict y values using the trained model
y_pred = model.predict(x_range)

# Plot the data points and the model's predictions
plt.scatter(X, y, color='blue', label='Data')
plt.plot(x_range, y_pred, color='red', label='SVR Model')
plt.xlabel('X')
plt.ylabel('y')
plt.title('SVR with Polynomial Kernel')
plt.legend()
plt.show()
```
