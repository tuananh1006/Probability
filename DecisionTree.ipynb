```python
from sklearn.datasets import load_iris
```

```python
data=load_iris()
```

```python
X=data['data']
y=data['target']
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(X,y)
```

```python
y_predict=model.predict(X)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y,y_predict)
```

```python
from sklearn.tree import export_graphviz
export_graphviz(model,out_file='tree.dot',feature_names=data['feature_names'],class_names=data['target_names'])
```

```python
import graphviz
with open("tree.dot") as f:
    dot_graph = f.read()
    graphviz.Source(dot_graph)
```

```python
graphviz.Source(dot_graph)
```

```python
model.predict_proba([[5,3,1,0.1]])
```

```python
import numpy as np
X=np.random.rand(100,1)*8
noise=np.random.randint(1,10,(100,1))
y=5+3*X+np.random.rand(100,1)+noise
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(X,y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
X_plot=np.arange(X.min(),X.max(),0.2).reshape(-1,1)
y_predict=model.predict(X_plot)
plt.plot(X_plot,y_predict,color='red')
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor(max_depth=2)
model.fit(X,y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
X_plot=np.arange(X.min(),X.max(),0.2).reshape(-1,1)
y_predict=model.predict(X_plot)
plt.plot(X_plot,y_predict,color='red')
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor(max_depth=8)
model.fit(X,y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
X_plot=np.arange(X.min(),X.max(),0.2).reshape(-1,1)
y_predict=model.predict(X_plot)
plt.plot(X_plot,y_predict,color='red')
```
